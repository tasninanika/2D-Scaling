```python
import matplotlib.pyplot as plt

# Step 1: Original square points
square = [(1, 1), (1, 4), (4, 4), (4, 1)]

# Step 2: Scaling factors
scale_x = 2  # make width 2 times
scale_y = 0.5  # make height half

# Step 3: Scale the square
scaled_square = []
for x, y in square:
    new_x = x * scale_x
    new_y = y * scale_y
    scaled_square.append((new_x, new_y))

# Step 4: Close the shapes
square.append(square[0])
scaled_square.append(scaled_square[0])

# Step 5: Separate x and y values for plotting
x1, y1 = zip(*square)           # original square
x2, y2 = zip(*scaled_square)    # scaled square

# Step 6: Plot both
plt.plot(x1, y1, 'go-', label='Original Square')
plt.plot(x2, y2, 'o-', label='Scaled Square')

# Extras
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Scaling of a Square")
plt.legend()
plt.axis('equal')
plt.show()
```
